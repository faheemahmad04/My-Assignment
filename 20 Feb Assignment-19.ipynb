Q1. Explain GET and POST methods.

Ans:- GET --- It is used by the browser to ask the server to send back the given resource.

POST --- It is the method in which the browser uses to talk to the server when asking for a response that takes into account, the data provided in the body of the HTTP request.

Q2. Why is request used in Flask?

Ans:- The request object is used to retrieve data from a form submitted with a POST request. It can also be used to retrieve data from a query string submitted with a GET request.

Q3. Why is redirect() used in Flask?

Ans:- Redirect() is used in the Flask to send the user to a particular URL with the status code. When we access a website, our browser sends a request to the server, and the server replies with what is known as the HTTP status code.
